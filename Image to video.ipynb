```python
imgs=[]
for i in range(1,51):
    img='image_'+str(i)+'.jpg'
    print(img)
    img=cv.imread(img)
    print(img.shape)
    imgs.append(img)
    #plt.imshow(img)
    
```

```python
plt.imshow(imgs[1])
```

```python
resized_img=[]
i=1
for img in imgs:
    print(i)
    image=cv.resize(img,(400,400))
    resized_img.append(image)
    i+=1
```

```python
height,width,layers=resized_img[0].shape
print(height,width,layers)
```

```python
video=cv.VideoWriter('ms_dhoni1.mp4',
                    cv.VideoWriter_fourcc(*'mp4v'),3,(width,height))
num_frames=50*3
repeated_images = np.tile(resized_img, (num_frames // len(imgs), 1, 1,1 ))
for img in repeated_images:    
    video.write(img)
    #print(i)
    #print(img)
cv.destroyAllWindows()
video.release()
```
